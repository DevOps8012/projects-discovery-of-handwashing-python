## Task 3

*...* 

```r
monthly <- read_csv("semmelweis_mortality_data/monthly_deaths.csv")

monthly$proportion_dead <- monthly$deaths / monthly$births
head(monthly)
```

## Task 4

*...* 

```r
ggplot(monthly, aes(date, proportion_dead)) +
  geom_point() + geom_line() +
  scale_y_continuous( limits = c(0, NA), labels = scales::percent)
```

## Task 5

*...* 

```r
handwashing_start <- as.Date("1847-06-01") # started in mid May, June was the first month with 100% handwashing
monthly$washing_hands <- monthly$date >= handwashing_start

ggplot(monthly_deaths, aes(date, proportion_dead, color = washing_hands)) +
  geom_point() + geom_line() +
  scale_y_continuous( limits = c(0, NA), labels = scales::percent)
```

## Task 6

*...* 

```r
prop_dead_before_washing <- monthly$proportion_dead[ ! monthly$washing_hands ]
prop_dead_after_washing <- monthly$proportion_dead[ monthly$washing_hands ]
mean(prop_dead_after_washing) - mean(prop_dead_before_washing)
```

## Task 7 (alternative 1)

*...* 

```r
t.test(prop_dead_after_washing, prop_dead_before_washing)
```

## Task 7 (alternative 2)

*...* 

```r
boot_difference <- c()
for(i in 1:1000) {
    boot_after_washing <- sample(prop_dead_after_washing, replace = TRUE)
    boot_before_washing <- sample(prop_dead_before_washing, replace = TRUE)
    boot_difference[i] <- mean(boot_before_washing) - mean(boot_after_washing)
}

quantile(boot_difference, c(0.025, 0.975))
```
